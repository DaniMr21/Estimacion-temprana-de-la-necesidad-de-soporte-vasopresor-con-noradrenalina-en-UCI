```python
#VARIABLES CON VARIANZA MUY BAJA

excluir = ['subject_id', 'hadm_id', 'stay_id',
           'horas_hasta_norad', 'etiqueta_norad_6_24']
num_cols = [c for c in df2.select_dtypes(include=[np.number]).columns
            if c not in excluir]

# Calculamos el coeficiente de variación (std/|mean|) para comparar entre escalas
resumen_var = pd.DataFrame({
    'std': df2[num_cols].std(),
    'mean': df2[num_cols].mean(),
})
resumen_var['cv'] = resumen_var['std'] / resumen_var['mean'].abs()
resumen_var.sort_values('cv').head(10)
```

```python
# Inspección de rangos para detectar posibles artefactos de medida
# que la winsorización no haya podido corregir.
rangos_esperados = {
    'anchor_age':        (18, 100),
    'hr_min':            (30, 200),    'hr_max':  (30, 250),
    'rr_min':            (5, 60),      'rr_max':  (5, 60),
    'map_min':           (20, 200),    'map_max': (20, 250),
    'temp_min':          (32, 42),     'temp_max': (32, 42),
    'spo2_min':          (50, 100),    'spo2_max': (50, 100),
    'fio2_min':          (21, 100),    'fio2_max': (21, 100),
    'ph_min':            (6.8, 7.7),   'ph_max':   (6.8, 7.7),
    'lactato_max':       (0, 25),
    'glucemia_max':      (30, 800),
    'diuresis_ml_kg_6h': (0, 40),
    'peso_kg':           (35, 250),
    'gcs_min':           (3, 15),      'gcs_max':  (3, 15),
}

anomalias = []
for col, (lo, hi) in rangos_esperados.items():
    if col in df2.columns:
        fuera = ((df2[col] < lo) | (df2[col] > hi)).sum()
        if fuera > 0:
            anomalias.append({
                'variable': col,
                'rango_esperado': f'[{lo}, {hi}]',
                'min_observado': df2[col].min(),
                'max_observado': df2[col].max(),
                'filas_fuera': fuera,
                'pct_fuera': round(100*fuera/len(df2), 2),
            })

pd.DataFrame(anomalias)
```

```python
# Verificamos que para cada triplete (min, media, max) se cumple la relación esperada.
# Después de winsorizar cada columna de forma independiente pueden aparecer inconsistencias.
grupos = ['lactato', 'creatinina', 'plaquetas', 'bilirrubina', 'tp',
          'gpt', 'got', 'pao2', 'ph', 'leucocitos', 'paco2',
          'bicarbonato', 'glucemia', 'hemoglobina',
          'hr', 'rr', 'temp', 'spo2', 'map', 'fio2', 'gcs']

inconsistencias = []
for g in grupos:
    m, mn, mx = f'{g}_media', f'{g}_min', f'{g}_max'
    if all(c in df2.columns for c in [m, mn, mx]):
        n_inc = ((df2[mn] > df2[m]) | (df2[m] > df2[mx])).sum()
        if n_inc > 0:
            inconsistencias.append({'grupo': g, 'filas_inconsistentes': n_inc})

if inconsistencias:
    print("Grupos con inconsistencias:")
    print(pd.DataFrame(inconsistencias).to_string(index=False))
else:
    print("No hay inconsistencias.")

total = sum(i['filas_inconsistentes'] for i in inconsistencias)
print(f"\nTotal de filas afectadas: {total} ({100*total/len(df2):.3f} %)")
```

```python
#¿Cuándo se inicia la noradrenalina dentro de la ventana 6-24h?
positivos = df2[df2['etiqueta_norad_6_24'] == 1]['horas_hasta_norad']

print(f"Estadísticos de horas_hasta_norad en positivos (n={len(positivos)}):")
print(positivos.describe().round(2).to_string())
```

```python
# Histograma de horas_hasta_norad en positivos
fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(positivos, bins=np.arange(6, 25, 1), color='#ff7f0e', alpha=0.8)
ax.axvline(positivos.median(), color='navy', linestyle='--',
           linewidth=2, label=f'Mediana = {positivos.median():.1f} h')
ax.axvline(positivos.mean(), color='darkred', linestyle=':',
           linewidth=2, label=f'Media = {positivos.mean():.1f} h')
ax.set_xlabel('Horas desde el ingreso en UCI hasta el inicio de noradrenalina')
ax.set_ylabel('Nº de estancias')
ax.set_title('Distribución temporal del evento en la clase positiva (n=356)')
ax.set_xticks(np.arange(6, 25, 2))
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# ¿Cómo se componen los 3.181 negativos?
neg = df2[df2['etiqueta_norad_6_24'] == 0]

neg_norad_tardia = neg[neg['horas_hasta_norad'].notna()]
neg_sin_norad   = neg[neg['horas_hasta_norad'].isna()]

print(f"Total de negativos: {len(neg)}")
print(f"  - Recibieron noradrenalina tras las 24 h ({neg_norad_tardia['horas_hasta_norad'].min():.1f}h a "
      f"{neg_norad_tardia['horas_hasta_norad'].max():.1f}h): {len(neg_norad_tardia)}")
print(f"  - No recibieron noradrenalina durante la estancia:              {len(neg_sin_norad)}")
```

```python
# Para los negativos que sí recibieron norad (pero tardíamente), ver cuándo la iniciaron
fig, ax = plt.subplots(figsize=(10, 5))
datos = neg_norad_tardia['horas_hasta_norad']
# Limitar el eje x por legibilidad (hay alguna estancia muy larga)
datos_mostrar = datos[datos <= 200]
ax.hist(datos_mostrar, bins=40, color='#1f77b4', alpha=0.7)
ax.axvline(24, color='red', linestyle='--', linewidth=2,
           label='Corte 24 h (límite de la ventana)')
ax.set_xlabel('Horas desde el ingreso en UCI hasta el inicio de noradrenalina')
ax.set_ylabel('Nº de estancias')
ax.set_title(f'Negativos con noradrenalina tardía (> 24 h). Se muestran '
             f'{len(datos_mostrar)} de {len(datos)} (resto >200h en la cola)')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Identificar pacientes que aparecen con etiqueta positiva en una estancia
# y negativa en otra. Esto justifica el uso de StratifiedGroupKFold en el CV.
por_paciente = df2.groupby('subject_id')['etiqueta_norad_6_24'].agg(['sum', 'count'])
pacientes_mixtos = por_paciente[(por_paciente['sum'] > 0) &
                                (por_paciente['sum'] < por_paciente['count'])]

print(f"Pacientes con etiqueta mixta entre sus estancias: {len(pacientes_mixtos)}")
print(f"  (positivo en al menos una estancia, negativo en al menos otra)\n")

if len(pacientes_mixtos) > 0:
    print("Desglose:")
    print(pacientes_mixtos.head(10).to_string())
```
